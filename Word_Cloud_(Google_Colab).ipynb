```python
# Step 3: Import libraries
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Step 5: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Step 6: Load the text file
with open("/content/drive/My Drive/JaneEyre.txt", "r") as file:
    text = file.read()
```

```python
# Step 7: Create a WordCloud object
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    min_font_size=10,
).generate(text)
```

```python
# Step 8: Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide axes
plt.tight_layout(pad=0)  # Adjust layout
plt.show()
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",
    stopwords=set(STOPWORDS),
    min_font_size=10,
    colormap='plasma',  # Example of a custom colormap
    max_words=200        # Example of limiting the number of words
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide axes
plt.tight_layout(pad=0)  # Adjust layout
plt.show()
```

```python
from PIL import Image
import numpy as np

# Load the mask image (make sure to replace with your actual path)
mask_image = np.array(Image.open("/content/drive/My Drive/heart.webp"))
plt.imshow(mask_image)
plt.axis("off")
plt.show()
```

```python
# Create the WordCloud object with a mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",
    mask=mask_image,
    stopwords=set(STOPWORDS),
    min_font_size=10,
).generate(text)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
